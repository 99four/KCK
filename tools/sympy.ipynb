```
from sympy.interactive import printing
printing.init_printing(use_latex=True)
import sympy as sym
# Definicje zmiennych
x, y = sym.symbols("x y")
```

```
eq1 = Eq(x**2 + 3*y - 10)
eq1
```

```
eq2 = Eq(4*x - y**2 + 2)
eq2
```

```
solve([eq1, eq2], x, y)
```
